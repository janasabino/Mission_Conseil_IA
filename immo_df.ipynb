# Importation des différentes bibliothèques:

Pandas pour manipuler et analyser de données. 
Numpy pour effectuer des calculs logiques et mathématiques sur des tableaux et des matrices

```python
import pandas as pd
import numpy as np
import re
```

# Importation du fichier dans un data frame

```python
immo_df = pd.read_csv("agence_immo.csv", sep=',', encoding = "utf-8")
immo_df.head()
```

# Etape de nettoyage


```python
immo_df['bedrooms']=immo_df['bedrooms'].str.extract(r'(\d+)')
immo_df['price']=immo_df['price'].str.replace("$","",regex=False)
immo_df.head()
```

# Changer le type d'un attribut pour un meilleur traitement

```python
immo_df['price']=immo_df['price'].astype(float)
immo_df.head()
```

# Créer d'autres dataframes à partir d'un dataframe selon nos besoins

```python
biens_df = immo_df[['bedrooms', 'sqft', 'city','id_proprio']].copy()
annonces_df = immo_df[['link', 'price', 'date']].copy()
biens_df
```

# Sauvegarder de notre Dataframe dans un fichier CSV

```python
biens_df.to_csv("biens.csv",index=False,encoding="utf-8")
annonces_df.to_csv("annonces.csv",index=False,encoding="utf-8")
```
