```python
import pandas as pd
import numpy as np
import re
```

```python
locations = pd.read_csv("locations.csv", sep=',', encoding = "utf-8")
locations
```

```python
def clean_bedrooms(df):
    df['bedrooms']=df['bedrooms'].str.extract(r'(\d+)')
    return df
clean_bedrooms(locations)
```

```python
def clean_surface(df, column):
   df[column] = df[column].str.findall(("\d+")).str[-1]
   df[column].fillna(0,inplace=True)
   df[column] = df[column].astype(int)
   df[column] = df[column].apply(lambda x :x if x>100 else None)
   return df[column]

locations["sqft"]=clean_surface(locations, "sqft")
locations
```

```python
locations['city'] = pd.Series(dtype='str')
locations
```

```python
mylist=['Paris','Lyon','Annecy','Boulogne-Billancourt','Aix-en-Provence','Bordeaux','Nice','Cannes','Saint-Tropez','Versailles','Levallois-Perret','Saint-Jean-Cap-Ferrat','Neuilly-sur-Seine','Ramatuelle','Saint-Mandé','Vincennes','Issy-les-Moulineaux','Suresnes','Saint-Cloud','']
locations['city'] = np.resize(mylist,len(locations))
locations
```

```python
del locations ["neighborhood"]
del locations['title']
del locations['pid']
del locations['date']
del locations['subregion']
```
